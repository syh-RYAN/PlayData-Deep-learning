### DataFrame의 컬럼 데이터 셋 Access

```python
# 1) titanic_df에 "Age_0" 열을 생성하고 모든 값을 0으로 설정하라
```

```python
titanic_df['Age_0'] = 0

titanic_df['Age_0']
```

```python
# 2) titanic_df에 "Age_by_10" 열을 생성하고 모든 값에 titanic_df["Age"] 열 데이터에 10을 곱하여 각각 할당하라
```

```python
titanic_df['Age_by_10'] = titanic_df['Age'] * 10

titanic_df['Age_by_10']
```

```python
# 3) titanic_df에 "Family_No" 열을 생성하고 
# 모든 값에 titanic_df["SibSp"] 열 데이터와 titanic_df["ParCh'] 열 데이터를 합하고 추가로 1을 더해서 할당하라
```

```python
titanic_df['Family_No'] = titanic_df["SibSp"] + titanic_df["Parch"] +1

titanic_df['Family_No']
```

```python
# 4) titanic_df에 "Age_by_10" 열의 모든 데이터에 titanic_df의 "Age_by_10"열 값에 10을 더하여 할당하라

titanic_df["Age_by_10"] = titanic_df["Age_by_10"] +10

titanic_df["Age_by_10"] 
```

* Aggregation 함수 적용

* groupby() 이용하기

```python
# 10) titanic_df의 선실 등급별 승객수를 groupby를 활용하여 조회하라
```

```python
titanic_df.pivot_table(index='Pclass',aggfunc='count')
```

```python
# 12) titanic_df의 선실 등급별로 총 승객수, 생존자수를 조회하라(groupby, pivot_table)
# groupby
titanic_df.groupby('Pclass')[['PassengerId','Survived']].agg({"PassengerId":"count","Survived":'sum'})
```

```python
# 위와 동일 결과 pivot_table
titanic_df.pivot_table(index='Pclass',
                       values = ['PassengerId','Survived'],
                       aggfunc={'PassengerId':'count','Survived':'sum'})
```

```python
# 타이타닉 성별의 데이터를 받아 'gender_person'열을 추가하라

g = titanic_df['Sex']

titanic_df['gender_person'] = g.apply(lambda x : x+'_person' )

titanic_df
```

```python
# 타이타닉 'Age'열의 정보 중 18세 이상 22세 이하이면 'Young'의 값을 가지고, 아니면 'Not young' 데이터를 가지는
# 'Age_Cat"열을 생성

c = titanic_df['Age']

titanic_df['Age_cat'] = c.apply(lambda c : 'Young' if(18  <= c <= 22) else 'Not Young')
titanic_df
```
